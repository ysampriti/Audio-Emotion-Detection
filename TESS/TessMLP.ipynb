```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.load("./DataX.npy")
Y = np.load("./DataY.npy")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import KFold, cross_validate
from sklearn.model_selection import validation_curve
```

```python
X_train, X_test, Y_train, Y_test  = train_test_split(X, Y,test_size=0.10,random_state=3)
```

```python
print((X_train.shape[0], X_test.shape[0]))
```

```python
model=MLPClassifier(alpha=0.01, batch_size=264, epsilon=1e-08,
                    hidden_layer_sizes=(300,), learning_rate='adaptive', max_iter=700, 
                    activation='tanh',learning_rate_init=0.001, verbose=True, early_stopping=True ,validation_fraction=0.15)
```

```python
# cvresults=cross_validate(model, X_train, Y_train, cv=10, return_train_score=True)
```

```python
# cvresults
```

```python
model.fit(X_train, Y_train)
```

```python
model.validation_scores_
```

```python
plt.plot(model.loss_curve_, label="loss")
plt.plot(model.validation_scores_,label="Validation_score")
plt.legend()
plt.show()
```

```python
vc = validation_curve(model,X_train,Y_train,param_name="alpha",param_range=[0.001,0.01],cv=10)
```
